```python
# Pandas
import pandas as pd
import psycopg2
import matplotlib.pyplot as plt
import scipy.stats as sts

# SQL Alchemy
from sqlalchemy import create_engine
```

```python
# Create Engine for homework (password is missing on purpose)
engine = create_engine('postgresql://postgres:<password>@localhost:5432/Homework')
connection = engine.connect()
```

```python
# Query All Records in the the salaries Table
salary_data = pd.read_sql("SELECT * FROM salaries", connection)

# Select salaries coloumn
salary = salary_data["salary"]
```

```python
# Create a histogram to visualize the most common salary ranges for employees.
plt.hist(salary)
plt.xlabel('Salary')
plt.ylabel('Counts')
plt.show()
print(sts.normaltest(salary.sample(50)))
```

```python
# Query All Records in the the salaries and join to title Table
title_salary_data = pd.read_sql("SELECT t.title, s.salary FROM employees e JOIN salaries s ON e.emp_no = s.emp_no JOIN titles t ON e.emp_title_id = t.title_id", connection)
```

```python
# Create a bar chart of average salary by title.

# Group by title
group_title = title_salary_data.groupby('title')['salary'].mean()

# Plot bar
group_title.plot(kind="bar", figsize=(7,5))

# Set a titles for the chart
plt.title("Average Salary per Title")
plt.xlabel("Employee Title")
plt.ylabel("Average Salary")
plt.show()
```
